```python
import tkinter as tk

window = tk.Tk()
window.title("그림판")
window.geometry("640x400")

#만약 여기서 6번째 줄이 주석처리된다면, 지금 생성한 frame은 크기 설정이 비어있기 때문에 최소한의 크기로 표시된다.
paint_frame = tk.Frame(window) 
paint_frame.grid(row=0, column=0, sticky=tk.W)

#그리기 버튼
#row나 column은 순서대로 배치됨
#Button을 만들 때 text가 들어가야 한다면 text 속성을 이용
paint_btn = tk.Button(paint_frame, text="pen", width=3)
#0행 0열에 위치시킴.
paint_btn.grid(row=0, column=0, sticky=tk.W)

#선그리기 버튼
line_btn = tk.Button(paint_frame, text="─", width=3)
line_btn.grid(row=0, column=1, sticky=tk.W)

#생성된 창이 유지되도록 하려면 mainloop 메소드를 실행
window.mainloop()
```

### tkinter.TK() 관련
+ window 창 생성하기
    + tkinter.TK()
+ 생성된 창이 유지되도록 하기
    + 창_이름.mainloop()
+ 생성한 창의 이름을 결정
    + 창_이름.title()
+ 창 너비, 높이, 생성 위치 설정
    + 창_이름.geometry()
    
### tkinter.Frame() 관련
+ 윈도우 창에 프레임 생성(생성한 프레임에 위젯(기능 or 부품)들을 추가할 것임)
    + tkinter.Frame(창_이름)
+ 프레임에 위젯을 배치시키는 방법 중 하나. 셀 단위로 위젯을 배치하는 함수.
    + 프레임_이름.grid()
        + row, column : 해당 행, 열로 위젯을 위치시킴
        + rowspan, columnspan : 현재 배치된 구역에서 위치를 조정
        + sticky : 현재 배치된 구역 안에서 특정 위치로 이동시킴.
            + tkinter.방향 : 방향 부분에 N, S, E, W, NE, NS, ...., NSEW 등을 적어줌(북, 남, 동, 서)

```python
import tkinter

window = tkinter.Tk()
window.title("그림판")
window.geometry("640x400")

color_list = ["white", "black", "red", "green", "blue", "cyan", "yellow"]

def bg_color() : 
    cs = listbox.get(listbox.curselection())
    window.config(bg = cs)

#리스트박스 부분
listbox = tkinter.Listbox(window)
for i, color in enumerate(color_list) : 
    listbox.insert(i, color)   
listbox.pack()

button = tkinter.Button(window, command = bg_color, width = 3, height = 1)
button.pack()

menubar=tkinter.Menu(window)
main_menu=tkinter.Menu(menubar, tearoff=1)
main_menu.add_command(label="New")
main_menu.add_separator()
main_menu.add_command(label="Exit", command=window.destroy)
menubar.add_cascade(label="File", menu=main_menu)
window.config(menu=menubar)

window.mainloop() 
```

```python
import tkinter as tk

window = tk.Tk()
window.title("그림판")
window.geometry("640x400")

draw_mode = 0
draw_color = "black"
x1 = 0
y1 = 0

"""
버튼 함수
"""
def lineButton():
    global draw_mode
    draw_mode = 1
    print(draw_mode)
def rectangleButton():
    global draw_mode
    draw_mode = 2
    print(draw_mode)
def circleButton():
    global draw_mode
    draw_mode = 3
    print(draw_mode)
def textButton():
    global draw_mode
    draw_mode = 4
    print(draw_mode)
def blackButton():
    global draw_color
    draw_color = "black"
    print(draw_color)
def whiteButton():
    global draw_color
    draw_color = "white"
    print(draw_color)
def redButton():
    global draw_color
    draw_color = "red"
    print(draw_color)
def orangeButton():
    global draw_color
    draw_color="orange"
    print(draw_color)
def yellowButton():
    global draw_color
    draw_color = "yellow"
    print(draw_color)
def greenButton():
    global draw_color
    draw_color = "green"
    print(draw_color)
def blueButton():
    global draw_color
    draw_color = "blue"
    print(draw_color)
def navyButton():
    global draw_color
    draw_color = "navy"
    print(draw_color)
def purpleButton():
    global draw_color
    draw_color = "purple"
    print(draw_color)

"""
그리기 도구 위젯
"""
style_frame = tk.Frame(window)
style_frame.grid(row=0, column=0, sticky="w")

line_button = tk.Button(style_frame, text="─", width=3, command=lineButton)
line_button.grid(row=0, column=0)

rectangle_button = tk.Button(style_frame, text="□", width=3, command=rectangleButton)
rectangle_button.grid(row=0, column=1)

circle_button = tk.Button(style_frame, text="○", width=3, command=circleButton)
circle_button.grid(row=0, column=2)

text_button = tk.Button(style_frame, text="text", width=3, command=textButton)
text_button.grid(row=0, column=3)

input_text = tk.StringVar()
input_entry = tk.Entry(style_frame, textvariable=input_text, width=25)
input_entry.grid(row=0, column=4)

"""
색상 위젯
"""
color_frame = tk.Frame(window)
color_frame.grid(row=0, column=1, sticky="e")

black_button = tk.Button(color_frame, bg="black", width=3, command=blackButton)
black_button.grid(row=0, column=0)

white_button = tk.Button(color_frame, bg="white", width=3, command=whiteButton)
white_button.grid(row=0, column=1)

red_button = tk.Button(color_frame, bg="red", width=3, command=redButton)
red_button.grid(row=0, column=2)

orange_button = tk.Button(color_frame, bg="orange", width=3, command=orangeButton)
orange_button.grid(row=0, column=3)

yellow_button = tk.Button(color_frame, bg="yellow", width=3, command=yellowButton)
yellow_button.grid(row=0, column=4)

green_button = tk.Button(color_frame, bg="green", width=3, command=greenButton)
green_button.grid(row=0, column=5)

blue_button = tk.Button(color_frame, bg="blue", width=3, command=blueButton)
blue_button.grid(row=0, column=6)

navy_button = tk.Button(color_frame, bg="navy", width=3, command=navyButton)
navy_button.grid(row=0, column=7)

purple_button = tk.Button(color_frame, bg="purple", width=3, command=purpleButton)
purple_button.grid(row=0, column=8)

"""
캔버스 위젯
"""
canvas_frame = tk.Frame(window, bd=2, bg="black")
canvas_frame.grid(row=2, column=0, columnspan=2, sticky="w")

canvas = tk.Canvas(canvas_frame, width=630, height=365, bg="white")
canvas.pack()

"""
마우스 이벤트 
"""
def mouseLDown(event):
    global x1, y1
    print("x :", event.x, ", y :", event.y)
    
    if draw_mode == 4:
        canvas.create_text(event.x, event.y, fill=draw_color, text=input_text.get())
        
    elif draw_mode >= 1 and draw_mode <= 3:
        x1 = event.x
        y1 = event.y

def mouseLUp(event):
    print("x1 :", x1, ", y1 :", y1, ", x2 :", event.x, ", y2 :", event.y)

    if draw_mode == 1:
        canvas.create_line(x1, y1, event.x, event.y, fill=draw_color, width=2)
        
    elif draw_mode == 2:
        canvas.create_rectangle(x1, y1, event.x, event.y, outline=draw_color, width=2)

    elif draw_mode == 3:
        canvas.create_oval(x1, y1, event.x, event.y, outline=draw_color, width=2)
    
canvas.bind("<Button-1>", mouseLDown)
canvas.bind("<ButtonRelease-1>", mouseLUp)

window.mainloop() 
```
