```python
import tkinter as tk

window = tk.Tk()
window.title("그림판")
window.geometry("640x400")

#만약 여기서 6번째 줄이 주석처리된다면, 지금 생성한 frame은 크기 설정이 비어있기 때문에 최소한의 크기로 표시된다.
paint_frame = tk.Frame(window) 
paint_frame.grid(row=0, column=0, sticky=tk.W)

#그리기 버튼
#row나 column은 순서대로 배치됨
#Button을 만들 때 text가 들어가야 한다면 text 속성을 이용
paint_btn = tk.Button(paint_frame, text="pen", width=3)
#0행 0열에 위치시킴.
paint_btn.grid(row=0, column=0, sticky=tk.W)

#선그리기 버튼
line_btn = tk.Button(paint_frame, text="─", width=3)
line_btn.grid(row=0, column=1, sticky=tk.W)

#생성된 창이 유지되도록 하려면 mainloop 메소드를 실행
window.mainloop()
```

### tkinter.TK() 관련
+ window 창 생성하기
    + tkinter.TK()
+ 생성된 창이 유지되도록 하기
    + 창_이름.mainloop()
+ 생성한 창의 이름을 결정
    + 창_이름.title()
+ 창 너비, 높이, 생성 위치 설정
    + 창_이름.geometry()
    
### tkinter.Frame() 관련
+ 윈도우 창에 프레임 생성(생성한 프레임에 위젯(기능 or 부품)들을 추가할 것임)
    + tkinter.Frame(창_이름)
+ 프레임에 위젯을 배치시키는 방법 중 하나. 셀 단위로 위젯을 배치하는 함수.
    + 프레임_이름.grid()
        + row, column : 해당 행, 열로 위젯을 위치시킴
        + rowspan, columnspan : 현재 배치된 구역에서 위치를 조정
        + sticky : 현재 배치된 구역 안에서 특정 위치로 이동시킴.
            + tkinter.방향 : 방향 부분에 N, S, E, W, NE, NS, ...., NSEW 등을 적어줌(북, 남, 동, 서)

```python
import tkinter

window = tkinter.Tk()
window.title("그림판")
window.geometry("640x400")

color_list = ["white", "black", "red", "green", "blue", "cyan", "yellow"]

def bg_color() : 
    cs = listbox.get(listbox.curselection())
    window.config(bg = cs)

#리스트박스 부분
listbox = tkinter.Listbox(window)
for i, color in enumerate(color_list) : 
    listbox.insert(i, color)   
listbox.pack()

button = tkinter.Button(window, command = bg_color, width = 3, height = 1)
button.pack()

menubar=tkinter.Menu(window)
main_menu=tkinter.Menu(menubar, tearoff=1)
main_menu.add_command(label="New")
main_menu.add_separator()
main_menu.add_command(label="Exit", command=window.destroy)
menubar.add_cascade(label="File", menu=main_menu)
window.config(menu=menubar)

window.mainloop() 
```

```python
#columnspan, rowspan 확인하기에 좋음

from tkinter import*

master=Tk()
master.title("String methods")

def cap():
    try:
        c=str.capitalize(s.get())
        s.set(c)
        
    except ValueError:
        pass

def cfold():
    try:
        c=str.casefold(s.get())
        s.set(c)
        
    except ValueError:
        pass

def ifier():
    try:
        if str.isidentifier(s.get())==True:
            s.set("Yes")
        
        else:
            s.set("No")
        
    except ValueError:
        pass

def ipace():
    try:
        if str.isspace(s.get())==True:
            s.set("Yes")
        
        else:
            s.set("No")

    except ValueError:
        pass

def iric():
    try:
       if str.isnumeric(s.get())==True:
           s.set("Yes")
       else:
           s.set("No")        
                
    except ValueError:
        pass

def ower():
    try:
        if str.islower(s.get())==True:
            s.set("Yes")
        
        else:
            s.set("No")        
        
    except ValueError:
        pass

def inum():
    try:
        if str.isalnum(s.get())==True:
            s.set("Yes")
        
        else:
            s.set("No")        
    except ValueError:
        pass

def imal():
    try:
        if str.isdecimal(s.get())==True:
            s.set("Yes")
        
        else:
            s.set("No")
        
    except ValueError:
        pass

    

def igit():
    try:
        if str.isdigit(s.get())==True:
            s.set("Yes")
        
        else:
            s.set("No")
        
    except ValueError:
        pass


def clr():
    try:
        s.set("")
        
              
    except ValueError:
        pass


s = StringVar()

entry=Entry(master, textvariable=s)
entry.grid(row=0, columnspan=4, sticky=W+E)


button_capitalize=Button(master, text="capitalize", width=10, command=cap)
button_capitalize.grid(row=2, column=0)

button_casefold=Button(master, text="casefold",width=10, command=cfold)
button_casefold.grid(row=2, column=1)

button_isidentifier=Button(master, text="isidentifier",width=10,command=ifier)
button_isidentifier.grid(row=2, column=2)


button_isspace=Button(master, text="isspace",width=10, command=ipace)
button_isspace.grid(row=3, column=0)

button_endswith=Button(master, text="endswith", width=10, command=iric)
button_endswith.grid(row=3, column=1)

button_islower=Button(master, text="islower",width=10, command=ower)
button_islower.grid(row=3, column=2)



button_isalnum=Button(master, text="isalnum",width=10, command=inum)
button_isalnum.grid(row=4, column=0)

button_isdecimal=Button(master, text="isdecimal",width=10, command=imal)
button_isdecimal.grid(row=4, column=1)

button_isdigit=Button(master, text="isdigit",width=10, command=igit)
button_isdigit.grid(row=4, column=2)

button_clear=Button(master, text="Clear", command=clr )
button_clear.grid(row=5, columnspan=3,sticky=W+E)
master.mainloop()
```
